USING KERAS

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])

y = np.array([[0],
              [1],
              [1],
              [0]])
```

```python
model = Sequential([
    Dense(8, activation='tanh', input_shape=(2,)),
    Dense(1, activation='sigmoid')
])
```

```python
model.compile(
    optimizer=Adam(learning_rate=0.1),
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
history = model.fit(
    X, y,
    epochs=1000,
    verbose=0
)
```

```python
loss, accuracy = model.evaluate(X, y, verbose=0)

print("Loss:", loss)
print("Accuracy:", accuracy)
```

```python
predictions = model.predict(X)
print("Predicted Outputs:")
print(np.round(predictions))
```

THIS IS USING PYTORCH

```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
X_torch = torch.tensor(X, dtype=torch.float32)
y_torch = torch.tensor(y, dtype=torch.float32)
```

```python
class XORNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.hidden = nn.Linear(2, 2)
        self.output = nn.Linear(2, 1)

    def forward(self, x):
        x = torch.tanh(self.hidden(x))
        x = torch.sigmoid(self.output(x))
        return x
model = XORNet()
```

```python
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.1)
```

```python
for epoch in range(500):
    optimizer.zero_grad()
    outputs = model(X_torch)
    loss = criterion(outputs, y_torch)
    loss.backward()
    optimizer.step()
```

```python
with torch.no_grad():
    outputs = model(X_torch)
    predictions = (outputs >= 0.5).float()
    accuracy = (predictions == y_torch).sum().item() / y_torch.size(0)

print("Accuracy:", accuracy)
print("Predictions:")
print(predictions)
```

USING TENSERFLOW

```python
import tensorflow as tf

X_tf = tf.constant(X, dtype=tf.float32)
y_tf = tf.constant(y, dtype=tf.float32)

W1 = tf.Variable(tf.random.normal([2,2]))
b1 = tf.Variable(tf.zeros([2]))

W2 = tf.Variable(tf.random.normal([2,1]))
b2 = tf.Variable(tf.zeros([1]))
```

```python
def model(x):
    h = tf.tanh(tf.matmul(x, W1) + b1)
    out = tf.sigmoid(tf.matmul(h, W2) + b2)
    return out
```

```python
loss_fn = tf.keras.losses.BinaryCrossentropy()
optimizer = tf.keras.optimizers.Adam(learning_rate=0.1)
```

```python
for epoch in range(1000):
    with tf.GradientTape() as tape:
        predictions = model(X_tf)
        loss = loss_fn(y_tf, predictions)

    gradients = tape.gradient(loss, [W1, b1, W2, b2])
    optimizer.apply_gradients(zip(gradients, [W1, b1, W2, b2]))
```

```python
predictions = model(X_tf)
predicted_classes = tf.cast(predictions >= 0.5, tf.float32)

accuracy = tf.reduce_mean(
    tf.cast(tf.equal(predicted_classes, y_tf), tf.float32)
)

print("Accuracy:", accuracy.numpy())
print("Predictions:")
print(predicted_classes.numpy())
```
