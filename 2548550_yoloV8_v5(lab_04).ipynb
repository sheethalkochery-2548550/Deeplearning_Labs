```python
import yaml

yaml_path = "/content/datasets/signature.yaml"

# load yaml
with open(yaml_path, "r") as f:
    data = yaml.safe_load(f)

# fix path for colab
data["path"] = "/content/datasets"

# save back
with open(yaml_path, "w") as f:
    yaml.dump(data, f)

print("✅ signature.yaml fixed")
```
